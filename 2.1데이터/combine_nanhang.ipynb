```python
import os
import zipfile
import pandas as pd

zip_dir = "/Users/Star1/Desktop/2025/2025á„‹á…§á„…á…³á†·/á„‰á…®á„ƒá…©á„€á…¯á†«_á„’á…¢á„á…¥á„á…©á†«/2.1á„‹á…§á†¯á„…á…µá†«á„ƒá…¦á„‹á…µá„á…¥á„€á…ªá†¼á„Œá…¡á†¼/data_crawling"         # ZIPë“¤ì´ ìžˆëŠ” í´ë”
parquet_dir = "/Users/Star1/Desktop/2025/2025á„‹á…§á„…á…³á†·/á„‰á…®á„ƒá…©á„€á…¯á†«_á„’á…¢á„á…¥á„á…©á†«/2.1á„‹á…§á†¯á„…á…µá†«á„ƒá…¦á„‹á…µá„á…¥á„€á…ªá†¼á„Œá…¡á†¼/parquets" # PARQUET ì €ìž¥ í´ë”
os.makedirs(parquet_dir, exist_ok=True)

errors = []

dtype_spec = {
    "o_admdong_cd": str,
    "d_admdong_cd": str,
    "st_time_cd": str,
    "fns_time_cd": str
}


for fname in os.listdir(zip_dir):
    if not fname.endswith(".zip"):
        continue

    zip_path = os.path.join(zip_dir, fname)
    base_name = fname.replace(".zip", "")
    csv_name = f"{base_name}.csv"
    parquet_path = os.path.join(parquet_dir, f"{base_name}.parquet")

    if os.path.exists(parquet_path):
        print(f"[SKIP] ì´ë¯¸ ì¡´ìž¬: {base_name}.parquet")
        continue

    try:
        with zipfile.ZipFile(zip_path, 'r') as zf:
            with zf.open(csv_name) as csvfile:
                df = pd.read_csv(csvfile, encoding="utf-8", dtype=dtype_spec, low_memory=False)  # í˜¹ì‹œ ê¹¨ì§€ë©´ cp949 ì‹œë„
                df.to_parquet(parquet_path, index=False)
                print(f"[OK] ë³€í™˜ ì™„ë£Œ: {base_name}.parquet")
    except Exception as e:
        print(f"[ERROR] {fname}: {e}")
        errors.append((fname, str(e)))

# ì—ëŸ¬ ìš”ì•½
if errors:
    print("\nâš ï¸ ë³€í™˜ ì‹¤íŒ¨ ëª©ë¡:")
    for f, msg in errors:
        print(f"{f} â†’ {msg}")
```

```python
# 1-3. í•©ì³ì§„ ë‚¨í–¥ë™ ìœ ì¶œì¸êµ¬ top4 ê·¸ë¦¬ê³  ì›”í‰ê·  ìˆ˜


import pandas as pd

df = pd.read_parquet("ë‚¨í–¥ë™_ì¶œë°œ_ì „ì²´.parquet")
df["st_time_cd"] = df["st_time_cd"].astype(int)
df["etl_ymd"] = pd.to_datetime(df["etl_ymd"].astype(str))

# 07:00~09:59 í•„í„°
valid_times = list(range(700, 920, 20))
df_filtered = df[df["st_time_cd"].isin(valid_times)]

# ì›” ë‹¨ìœ„ ì»¬ëŸ¼ ì¶”ê°€
df_filtered["month"] = df_filtered["etl_ymd"].dt.to_period("M")

# ë„ì°©ì§€ë³„ ì›”ë³„ ì´ ìœ ì¶œëŸ‰
monthly = (
    df_filtered
    .groupby(["d_admdong_cd", "month"])["total_cnt"]
    .sum()
    .reset_index()
)

# ë„ì°©ì§€ë³„ ì›”í‰ê·  ìœ ì¶œëŸ‰ ê³„ì‚°
monthly_avg = (
    monthly
    .groupby("d_admdong_cd")["total_cnt"]
    .mean()
    .reset_index()
    .rename(columns={"total_cnt": "monthly_avg_cnt"})
)

# ë„ì°©ì§€ë³„ ì „ì²´ ì´í•©ë„ ê°™ì´ êµ¬í•¨
total_sum = (
    df_filtered
    .groupby("d_admdong_cd")["total_cnt"]
    .sum()
    .reset_index()
    .rename(columns={"total_cnt": "total_cnt"})
)

# í•©ì¹˜ê¸°
result = pd.merge(total_sum, monthly_avg, on="d_admdong_cd")

# ìƒìœ„ 3ê°œ ì¶œë ¥
top3 = result.sort_values("total_cnt", ascending=False).head(4)

print("âœ… ë‚¨í–¥ë™ ì¶œë°œ (07~09ì‹œ) â†’ ë„ì°©ì§€ Top 4 ë° ì›”í‰ê·  ìœ ì¶œ ì¸êµ¬:")
print(top3)
```

```python
# 1-3. í•©ì³ì§„ ë‚¨í–¥ë™ ìœ ì¶œì¸êµ¬ top4 í‰ê· (ì¤‘ì•™ê°’) ì´ë™ì‹œê°„

# ë„ì°©ì§€ top3 â†’ ë¦¬ìŠ¤íŠ¸ë¡œ
top3_dests = [11620775.0, 11620765.0, 11620735.0]

# í•´ë‹¹ ë„ì°©ì§€ë“¤ë§Œ í•„í„°ë§
df_top3 = df[df["d_admdong_cd"].isin(top3_dests)]


# í†µê³„ ìš”ì•½
move_time_stats = (
    df_top3
    .groupby("d_admdong_cd")["move_time"]
    .agg(
        count="count",
        mean="mean",
        median="median",
        q1=lambda x: x.quantile(0.25),
        q3=lambda x: x.quantile(0.75),
        min="min",
        max="max"
    )
    .reset_index()
)

# ê²°ê³¼ ë³´ê¸°
print("âœ… ë„ì°©ì§€ë³„ ì´ë™ì‹œê°„ í†µê³„ ìš”ì•½ (ë‹¨ìœ„: ë¶„):")
print(move_time_stats)
```

```python
# 1. ë„ì°©ì§€ Top 4 ë¦¬ìŠ¤íŠ¸
top4_dests = [11620775.0, 11620765.0, 11620735.0, 11620745.0]

# 2. í•„í„°ë§
df_top = df_filtered[df_filtered["d_admdong_cd"].isin(top4_dests)].copy()

# 3. ì›” ì»¬ëŸ¼ ìƒì„±
df_top["month"] = df_top["etl_ymd"].dt.to_period("M")

# 4. ì—°ë ¹ëŒ€ ì»¬ëŸ¼
age_cols = ["0010_cnt", "2030_cnt", "4050_cnt", "60plus_cnt", "total_cnt"]

# 5. ì›”ë³„ í•©ê³„
monthly_age = (
    df_top
    .groupby(["d_admdong_cd", "month"])[age_cols]
    .sum()
    .reset_index()
)

# 6. ì›”í‰ê·  ê³„ì‚°
monthly_avg_age = (
    monthly_age
    .groupby("d_admdong_cd")[age_cols]
    .mean()
    .reset_index()
)

# 7. ë¹„ìœ¨ ê³„ì‚°
for col in ["0010_cnt", "2030_cnt", "4050_cnt", "60plus_cnt"]:
    monthly_avg_age[f"{col}_ratio"] = monthly_avg_age[col] / monthly_avg_age["total_cnt"]

# 8. ì¶œë ¥

age_ratio_cols = [
    "d_admdong_cd",
    "0010_cnt_ratio", "2030_cnt_ratio",
    "4050_cnt_ratio", "60plus_cnt_ratio"
]

print("âœ… ë„ì°©ì§€ Top 4 ê¸°ì¤€ ì›”í‰ê·  ìœ ì¶œ ì¸êµ¬ (ì—°ë ¹ëŒ€ë³„):")
print(monthly_avg_age[age_ratio_cols])
```
