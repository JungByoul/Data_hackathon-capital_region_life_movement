```python

from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os, time

# ✅ 다운로드 경로 설정
download_dir = "/Users/Star1/Desktop/2025/2025여름/수도권_해커톤/2.1열린데이터광장/data_crawling"  # 원하는 다운로드 경로 지정


chrome_options = Options()
chrome_options.add_experimental_option("prefs", {
    "download.default_directory": download_dir,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
})
driver = webdriver.Chrome(options=chrome_options)
wait = WebDriverWait(driver, 15)

# ✅ 페이지 진입
url = "https://data.seoul.go.kr/dataList/OA-22299/F/1/datasetView.do"
driver.get(url)

# ✅ '전체 파일보기' 클릭
wait.until(EC.element_to_be_clickable((By.XPATH, '//a[.//span[text()="전체 파일보기"]]'))).click()
time.sleep(5)  # 로딩 대기

# ✅ 모든 .zip span 태그 추출
all_spans = driver.find_elements(By.CSS_SELECTOR, 'td.align-l span[title$=".zip"]')
print(f"[INFO] Total files found: {len(all_spans)}")


# ✅ 152~517행 다운로드
for i, span in enumerate(all_spans[151:517], start=152):
    file_name = span.get_attribute("title")
    file_path = os.path.join(download_dir, file_name)
    if os.path.exists(file_path):
        print(f"[SKIP] {file_name}")
        continue
    print(f"[{i}] Downloading: {file_name}")
    try:
        span.click()
        time.sleep(2)
    except Exception as e:
        print(f"[ERROR] {file_name}: {e}")
```
